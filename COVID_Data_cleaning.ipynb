**Standardizing the "land_area_sq_km" column**

```python
df_population["land_area_sq_km"].unique()
```

```python
std_pop = pd.DataFrame(df_population["land_area_sq_km"])
```

```python
std_pop
```

```python
def standard_population(area):
    if area.endswith('M'):
        return float(area[:-1]) * 1000
    elif area.endswith('K'):
        return float(area[:-1])
    else:
        if "<" in area:
            return float(area.lstrip("< "))
        else:
            return float(area)
```

```python
std_pop["std_land_area"] = std_pop["land_area_sq_km"].apply(lambda x: standard_population(x))
```

```python
std_pop["std_land_area"].unique()
```
